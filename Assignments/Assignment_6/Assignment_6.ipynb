# Question - 4
Define following.

# Answer

1. # <u>Class:</u>
    * A class is a code template or blueprint for creating objects. Objects have member variables and have behaviour associated with them. In python a class is created by the keyword class . 

2. # <u>Object:</u>
    *  Object is simply a collection of data (variables) and methods (functions) that act on those data.

3. # <u>Attribute:</u>
    *  In computing, an attribute is a specification that defines a property of an object, element, or file. It may also refer to or set the specific value for a given instance of such.
    
4. # <u>Behaviour:</u>
    * A behaviour of an object is the things it can do which is stored in methods.

```python
class Car():
    def __init__(self,color,model,name,company,engine_power):
                                
                                #attributes
        self.color = color
        self.model = model
        self.name = name
        self.company = company
        self.engine_power = engine_power
        
                                #methods                
        
    def update_color(self,newcolor):
        self.color = newcolor
        
    def update_model(self,newmodel):
        self.model = newmodel
        
    def update_company(self,newcompany):
        self.company = newcompany
        
                                #instances

ins_1 = Car('Black','2019','Chiron','Bugatti','1479 Hp')
print('First Instance :')
print('\n')
print(ins_1.color)
print(ins_1.model)
print(ins_1.name)
print(ins_1.company)
print(ins_1.engine_power)
print('\n')

ins_1.update_color('Red')

print('Second Instance :')
print('\n')
print(ins_1.color)
print(ins_1.model)
print(ins_1.name)
print(ins_1.company)
print(ins_1.engine_power)
print('\n')

ins_1.update_model(1996)
print('Third Instance :')
print('\n')
print(ins_1.color)
print(ins_1.model)
print(ins_1.name)
print(ins_1.company)
print(ins_1.engine_power)
print('\n')

ins_1.update_company('McLaren')
print('Fourth Instance :')
print('\n')
print(ins_1.color)
print(ins_1.model)
print(ins_1.name)
print(ins_1.company)
print(ins_1.engine_power)
print('\n')

ins_1 = Car('Green','2020','A8','Audi','2000 Hp')
print('Fifth Instance :')
print('\n')
print(ins_1.color)
print(ins_1.model)
print(ins_1.name)
print(ins_1.company)
print(ins_1.engine_power)
print('\n')
```
